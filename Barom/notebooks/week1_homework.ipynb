## 요청사항3
필요한 열 선택 / ETL 처리

Q.ETL 처리?

```python
pre2_df = pre1_df.select(
    col('date'),
    col('ul.USER_ID_hash').alias('user_id'),
    col('ul.SEX_ID').alias('sex_id'),
    col('ul.PREF_NAME').alias('name'),
    col('cdt.SMALL_AREA_NAME').alias('area_name'),
    col('cdt.PURCHASEID_hash').alias('purchased_id'),
    col('cdt.ITEM_COUNT').alias('item_count')
)
```

```python
pre2_df.show()
```

## 요청사항4
결측치 처리
- 구분열 결측치 "-"으로, 수치열 결측치 0으로 처리
- "NA"로 된 값도 위에 정의한 값으로 결측치 처리

| memo
- isNull : null과 None 값만 결측치로 판단(대소문자 구분없이) / 만약 'NULL', 'Null', 'null'과 같은 문자열 값이 들어가면 당연히 결측치로 판단하지 않음

```python
pre2_df.printSchema()
```

```python
pre3_df = pre2_df.na.fill('-', subset=['date', 'user_id', 'sex_id', 'name', 'area_name', 'purchased_id']) \
    .na.fill(0, subset=['item_count'])
```

```python
pre3_df.show()
```

```python
# from pyspark.sql.functions import sum, when

missing_cnt = pre3_df.select([
    _sum(when(
        col(c).isNull() | (col(c) == "") | (col(c) == " ") | (col(c) == "NA"), 
        1).otherwise(0)).alias(c) 
    for c in pre3_df.columns
])
```

```python
missing_cnt.show()
```

## 요청사항5
date 컬럼을 활용해 year / month / day / dow / weeks 열 만들기
- dow의 경우 일요일이 0, 토요일이 6으로 표현

```python
from pyspark.sql.functions import year, month, dayofmonth, dayofweek, weekofyear
```

```python
pre4_df = pre3_df \
    .withColumn('year', year('date')) \
    .withColumn('month', month('date')) \
    .withColumn('day', dayofmonth('date')) \
    .withColumn('dow', (dayofweek('date') - 1)) \
    .withColumn('weeks', weekofyear('date'))
```

```python
columns = ['date', 'year', 'month', 'day', 'dow', 'weeks', 'user_id', 'sex_id', 'name', 'area_name', 'purchased_id', 'item_count']
pre4_df = pre4_df.select(columns)
```

```python
pre4_df.show()
```

## 요청사항6
구분열 기준 수치열 합으로 중복 제거

Q.년, 월, 일, 요일, 주차 컬럼은 수치열로 보는 것이 아닌게 맞는지?

| memo

sum을 _sum으로 사용하는 이유?

python의 내장함수 sum과 충돌하기 때문
- python 내장함수 sum : iterable의 합계
- pyspark sql 함수 sum : DataFrame의 열에 대해 합계

```python
pre4_df.printSchema()
```

```python
# from pyspark.sql.functions import sum as _sum

pre5_df = pre4_df.groupBy('date', 'year', 'month', 'day', 'dow', 'weeks', 'user_id', 'sex_id', 'name', 'area_name', 'purchased_id').agg(
    _sum('item_count').alias('item_count')
)
```

```python
pre5_df.show()
```
